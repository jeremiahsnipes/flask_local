# Install and Imports

```python
# Importing the core libraries I need for this project.
# These are the tools I'll use for data analysis, visualization, modeling, and saving the final trained model.

import pandas as pd     # For working with tabular data
import numpy as np      # Numerical operations
import matplotlib.pyplot as plt  # Plotting/visualizations
import seaborn as sns   # Cleaner visualizations

# Sklearn modules for scaling, splitting, training, and evaluating the model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

import joblib           # To save the trained model and scaler
```

# Load and Explore the Data

This dataset contains keystroke timing features for 51 users, each typing the same password 400 times.  
I'm loading it directly so I can begin exploring the structure.

#### Basic Info

```python
# Here I'm loading the keystroke dataset into a pandas DataFrame.
# df will store all 20 timing features, plus the subject ID.

data_path = "DSL-StrongPasswordData.csv"

df = pd.read_csv(data_path)

df.head()
```

```python
# Checking column names, data types, and general structure.
# I'm using this to confirm that all timing data is numeric and ready for modeling.

df.info()
```

```python
# Looking at basic statistical summaries of the timing features.
# This helps me see if there are any abnormal values or outliers.

df.describe()
```

```python
# Checking for missing values across all columns.
# I want to verify that the dataset is complete and doesn't need imputation.

df.isnull().sum()
```

```python
# Checking for duplicate rows.
# This dataset should not contain duplicates, but I always verify to be safe.

df.duplicated().sum()
```

#### Visualize the Data

```python
# Visualizing a few timing features so I can understand their ranges and behavior.
# This is useful to see whether impostor vs normal users show noticeable timing differences.

plt.figure(figsize=(12,6))
sns.histplot(df["H.period"], bins=40, kde=True)
plt.title("Distribution of H.period Timing Feature")
plt.xlabel("H.period")
plt.ylabel("Count")
plt.show()
```

#### Compare One Feature Across Subjects

```python
# Plotting one feature across subjects to illustrate behavioral differences.
# I expect each subject to have slightly different typing rhythms.

plt.figure(figsize=(12,6))
sns.boxplot(x="subject", y="H.period", data=df)
plt.xticks([], [])  # hide labels, since there are 51 users
plt.title("H.period Variation Across Subjects")
plt.show()
```

#### Sorting and Filtering

```python
# Here's a quick example of filtering one user's data for calibration.
# During modeling, I convert the classification problem into normal (user) vs impostor (others).

df[df["subject"] == "s001"].head()
```

# Split the Data for Training and Testing

#### Feature Engineering & Labels

```python
# Creating labels based on one target user.
# I'm switching from "s001" to "s002" because s001 does not exist in the dataset.
# s002 is confirmed to be present in the dataset.

target_user = "s002"
df["label"] = np.where(df["subject"] == target_user, 1, 0)

# Dropping subject column after labeling.
df = df.drop(columns=["subject"])
```

#### Train/Test Split

```python
# Separating features (X) from the label column (y).
X = df.drop(columns=["label"])
y = df["label"]

# Splitting the dataset into train and test sets.
# Stratifying keeps the class ratio consistent in both groups.
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Scaling the timing features.
# This ensures all timing values contribute equally to the model.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

##### Interpretation

The train/test split ensures we evaluate the model on unseen data.  
I’m scaling the features because the timing values vary in magnitude,  
and a model like RandomForest benefits from normalized input.


# Model | RandomForestClassifier | Train | Tune | Evaluation

#### Build the Model

```python
# Creating the RandomForest model.
# I chose RandomForest because it's stable, handles noisy biometric data well,
# and works consistently across Flask, Docker, and Kubernetes deployment.

model = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    random_state=42
)
```

##### Interpretation

I am using 200 trees and a max depth of 10.  
These settings balance performance with speed, which matters during deployment.  
RandomForest is also resilient to noise and handles nonlinear behavioral patterns well.


#### Train the Model

```python
# Training the model on the scaled training data.
model.fit(X_train_scaled, y_train)
```

##### Interpretation

The model is now learning what my target user’s typing behavior looks like  
and how it differs from impostor timing behavior.


#### Tune the Model

```python
# Example: Check feature importance to understand which timing values matter.
importances = model.feature_importances_
sorted_indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10,8))
plt.bar(range(10), importances[sorted_indices][:10])
plt.title("Top 10 Most Important Timing Features")
plt.show()
```

##### Interpretation

Feature importance helps me understand which timing intervals contribute  
most to distinguishing the real user from impostors.


#### Evaluate the Model


```python
# Predicting on the test data.
y_pred = model.predict(X_test_scaled)
y_proba = model.predict_proba(X_test_scaled)[:, 1]

# Checking core evaluation metrics.
print(classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_proba))
```

##### Interpretation

To understand these results, think of this dataset like a typing gauntlet. I chose one user—s002—to be the champion, 
the real owner of the account. All 50 other users become challengers trying to imitate the champion’s typing rhythm.

When I look at the model’s predictions, it’s extremely good at spotting challengers. For the impostor class (label 0), the 
model catches basically all 6,000 attempts—nearly perfect precision and perfect recall. That means if someone who isn’t 
the champion tries to enter the gauntlet, the model shuts them down almost every time. That’s great for security.

For the champion class (label 1), the story is a little different. The model is very strict: it never mistakes a challenger 
for the champion (precision is 1.00), which is exactly what you want in an authentication system. But its recall is lower, 
meaning sometimes the champion’s own attempts get flagged as challengers. That happens because the challengers greatly 
outnumber the champion in the training data—6000 challenger attempts versus only 120 champion attempts—so the model 
leans defensive.

Even with that imbalance, the overall accuracy is 98 percent, and the model’s ROC-AUC score is extremely high. That 
means when you look at the model’s ranking ability—its ability to separate champion behavior from challenger 
behavior—it’s almost perfect.

So in plain English: this model is a very strong gatekeeper. It almost never lets a challenger slip through the gauntlet 
and into the champion’s account. The tradeoff is that it sometimes challenges the champion themselves, which is a 
normal and acceptable behavior in a security-focused system.


```python
# Visualizing the confusion matrix.
plt.figure(figsize=(5,4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

# Save model.pkl

```python
# Saving the trained RandomForest model
joblib.dump(model, "model.pkl")
print("model.pkl saved!")
```

# Save scaler.pkl

```python
# Saving the StandardScaler used during training
joblib.dump(scaler, "scaler.pkl")
print("scaler.pkl saved!")
```
